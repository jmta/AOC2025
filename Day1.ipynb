```python
from aocd.models import Puzzle
import math
puzzle = Puzzle(year=2025, day=1)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
puzzle.answer_b = partTwo(puzzle.input_data)
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b): #Slight issue on this one, hence the line above, the answer_b on example is not correct
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")

```

```python
def partOne(input):
    lines = input.split()
    dial = 50
    password = 0
    for line in lines:
        if line[0] == 'L':
            dial -= int(line[1:])
        else:
            dial += int(line[1:])

        if dial%100 == 0:
           password += 1
    return password
```

```python
def partTwo(input):
    lines = input.split()
    dial = 50
    password = 0
    for line in lines:
        before = dial
        spin = int(line[1:])
        while spin > 0:
            if line[0] == 'L':
                dial -= 1
            else:
                dial += 1
    
            if dial%100 == 0:
               password += 1
            spin -= 1
    return password
```
