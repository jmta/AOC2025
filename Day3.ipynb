```python
from aocd.models import Puzzle
import math
import textwrap
puzzle = Puzzle(year=2025, day=3)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b): 
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")

```

```python
def partOne(input):
    lines = input.split()
    ans = 0
    for line in lines:
        #First find the largest number in the string (not including the last digit)
        location = 0
        firstValue = 0
        for i in range(0, len(line)-1):
            numberAtPosition = int(line[i])
            if numberAtPosition > int(firstValue):
                firstValue = line[i]
                location = i

        secondValue = 0
        for i in range(location+1, len(line)):
            numberAtPosition = int(line[i])
            if numberAtPosition > int(secondValue):
                secondValue = line[i]
        ans += int(firstValue+secondValue)
    return ans
```

```python
def partTwo(input):
    lines = input.split()
    joltageLength = 12
    ans = 0
    for line in lines:
        #First find the largest number in the string (not including the last digit)
        location = 0
        firstValue = 0
        lineJoltage = []
        lastLocation = -1
        for j in list(reversed(range(0,joltageLength))):
            value = 0
            for i in range(lastLocation+1, len(line)-j):
                numberAtPosition = int(line[i])
                if numberAtPosition > int(value):
                    value = line[i]
                    lastLocation = i
            lineJoltage.append(value)
        ans += int(''.join(lineJoltage))
        
    return ans
```
