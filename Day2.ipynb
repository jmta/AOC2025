```python
from aocd.models import Puzzle
import math
import textwrap
puzzle = Puzzle(year=2025, day=2)
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a):
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2

if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b): 
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")

```

```python
def partOne(input):
    ranges = input.split(',')
    ret = 0
    for rang in ranges:
        lower = rang.split('-')[0]
        upper = rang.split('-')[1]
        for number in range(int(lower), int(upper)+1):
           if stringRepeatsHalf(str(number)):
               ret += number
    return ret
```

```python
def partTwo(input):
    ranges = input.split(',')
    ret = 0
    for rang in ranges:
        lower = rang.split('-')[0]
        upper = rang.split('-')[1]
        for number in range(int(lower), int(upper)+1):
           if stringRepeats(str(number)):
               ret += number
    print(ret)
    return ret
```

```python
def stringRepeatsHalf(s1):
    s2, s3 = s1[:len(s1)//2 + len(s1)%2], s1[len(s1)//2 + len(s1)%2:]
    return s2 == s3
```

```python
def stringRepeats(s1):
    for split in range(1,math.floor(len(s1)/2)+1):
        if(len(set(textwrap.wrap(s1, split))) == 1):
            return True
    return False
        
```
