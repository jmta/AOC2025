```python
from aocd.models import Puzzle
puzzle = Puzzle(year=2025, day=5)
printDebugOutput = False
```

```python
#Solve Part 1
if partOne(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_a): #answer isn't correct on this one...
    print("Think we have an answer for part one")
    puzzle.answer_a = partOne(puzzle.input_data)
else:
    print("Incorrect for part 1")

    
#Solve Part 2
if partTwo(puzzle.examples[0].input_data) == int(puzzle.examples[0].answer_b): 
    print("Think we have an answer for part two")
    puzzle.answer_b = partTwo(puzzle.input_data)
else:
    print("Incorrect for part 2")

```

```python
def partOne(input):
    freshIngredients = []
    answer = 0
    for line in input.split():
        if '-' not in line:
            fresh = False
            for ranges in freshIngredients:
                split = ranges.split('-')
                if int(line) >= int(split[0]) and int(line) <= int(split[1]):
                    fresh = True

            if fresh:
                answer += 1
        else:
            freshIngredients.append(line)
        
    return answer
```

```python
def partTwo(input):
    ranges = []

    for line in input.split():
        if '-' in line:
            lower = int(line.split('-')[0])
            upper = int(line.split('-')[1])
            ranges.append([lower,upper])

    ranges.sort()
 
    newRanges = []
    for query in ranges:
        if len(newRanges) == 0:
            newRanges.append(query)
        else:
            foundInRanges = False
            for rang in newRanges:
                #First check we aren't completely in the range
                if safeIsInRange(query[0], rang[0], rang[1]) and safeIsInRange(query[1], rang[0], rang[1]):
                    #Don't adjust any ranges, but we don't want to add this range
                    #print(query + rang)
                    foundInRanges = True
                elif safeIsInRange(query[0], rang[0], rang[1]):
                    #print("Replaced upper " + str(rang[1]) + " with " + str(query[1]))
                    rang[1] = query[1]
                    foundInRanges = True
                elif safeIsInRange(query[1], rang[0], rang[1]):
                    #print("Replaced lower " + str(rang[0]) + " with " + str(query[0]))
                    rang[0] = query[0]
                    foundInRanges = True
            if not foundInRanges:
                newRanges.append(query)

    ranges = newRanges
    
    #We should now have all ranges, so need to count how many in each range
    answer = 0
    for rang in ranges:
        if printDebugOutput:
            print("Range: " + str(rang[0]) + ", " + str(rang[1]))
        answer += rang[1] - rang[0] + 1

    return answer
```

```python
def safeIsInRange(query, lower, upper):
    return query >= lower and query <= upper
```
